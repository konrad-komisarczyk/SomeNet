# Porównanie szybkości uczenia dla różnych wartości batch_size

Przeprowadzę eksperyment dla zbioru `steps_small` dla dokładnie tej samej architektury i parametrów co we wczesniej przedstawionym przykładzie. Wyrysuję wykresy MSE po kolejnych aplikacjach zmiany wag w czasie (w zalęzności od liczby przedstawionych przykładów).

```python
def plot_mses_by_batch_size(batch_size, n_epochs=1000, from_epoch=50):
    net.add(mn.DenseLayer(15, "sigmoid"))
    net.add(mn.DenseLayer(15, "sigmoid"))
    net.add(mn.DenseLayer(1, "identity"))
    net.kernel_init("xavier")

    mse, iters, mses_history = net.backpropagate(steps_train_X, steps_train_Y, 
                               eta=0.00004, n_epochs=n_epochs,
                               required_mse=100, batch_size=batch_size, 
                               verbose=0,
                               validation_input=steps_test_X, 
                               validation_output=steps_test_Y
                              )
    if batch_size is None:
        from_iter = from_epoch
    else:
        from_iter = int(from_epoch * 50 / batch_size)
    
    iters, mses = zip(*mses_history[from_iter:])
    plt.plot(iters, mses, linewidth=0.25)
    
```

```python
plot_mses_by_batch_size(None)  # <=> batch_size=50
```

```python
plot_mses_by_batch_size(25)
```

```python
plot_mses_by_batch_size(10)
```

```python
plot_mses_by_batch_size(5)
```

```python
plot_mses_by_batch_size(1)
```

Widzimy, że mniejszy rozmiar batcha powoduje większą niestabilność wyniku.
