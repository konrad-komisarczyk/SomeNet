```python
import pandas as pd
import numpy as np
import my_net as mn
import matplotlib.pyplot as plt
```

```python
square_train = pd.read_csv('../data/regression/square-large-training.csv', index_col=0)
steps_train = pd.read_csv('../data/regression/steps-large-training.csv', index_col=0)
square_test = pd.read_csv('../data/regression/square-large-test.csv', index_col=0)
steps_test = pd.read_csv('../data/regression/steps-large-test.csv', index_col=0)
square_test_X = square_test[['x']].values
square_test_Y = square_test[['y']].values
steps_test_X = steps_test[['x']].values
steps_test_Y = steps_test[['y']].values
square_train_X = square_train[['x']].values
square_train_Y = square_train[['y']].values
steps_train_X = steps_train[['x']].values
steps_train_Y = steps_train[['y']].values

multimodal_train = pd.read_csv('../data/regression/multimodal-large-training.csv')
multimodal_test = pd.read_csv('../data/regression/multimodal-large-test.csv')
multimodal_test_X = multimodal_test[['x']].values
multimodal_test_Y = multimodal_test[['y']].values
multimodal_train_X = multimodal_train[['x']].values
multimodal_train_Y = multimodal_train[['y']].values
```

# square

## momentum

```python
np.random.seed(21)
net = mn.Net(1)
net.add(mn.DenseLayer(25, "sigmoid"))
net.add(mn.DenseLayer(25, "sigmoid"))
net.add(mn.DenseLayer(25, "sigmoid"))
net.add(mn.DenseLayer(1, "identity"))
net.kernel_init("xavier")

mse, iters, mses_history = net.backpropagate(square_train_X, square_train_Y, 
                               eta=0.000005, n_epochs=3000,
                               required_mse=1, batch_size=None, 
                               verbose=1,
                               validation_input=square_test_X, 
                               validation_output=square_test_Y, normalization_method="momentum"
                              )
```

```python
iters, mses = zip(*mses_history[10:])
plt.plot(iters, mses, linewidth=0.25)
```

Sie uczy si bardzo szybko, wystpuj po drodze czasami pewne perturbacje, po ktrych MSE znowu si stabilizuje.

```python
preds = net.predict(square_test_X)
plt.scatter(square_test_X, preds, linewidth=0.25)
```

Sie jest w stanie nauczy si w miar podobnego ksztatu do oczekiwanego, jednak MSE pozostaje na poziomie okoo 500. Inne noniki zbioru testowego i treningowego utrudniaj uzyskanie niskiego MSE na zbiorze testowym.

## RMSProp

```python
np.random.seed(21)
net = mn.Net(1)
net.add(mn.DenseLayer(25, "sigmoid"))
net.add(mn.DenseLayer(1, "identity"))
net.kernel_init("xavier")

mse, iters, mses_history = net.backpropagate(multimodal_train_X, multimodal_train_Y, 
                               eta=0.1, n_epochs=500,
                               required_mse=1, batch_size=None, 
                               verbose=1,
                               validation_input=multimodal_test_X, 
                               validation_output=multimodal_test_Y, normalization_method="RMSProp"
                              )
```

```python
iters, mses = zip(*mses_history[10:])
plt.plot(iters, mses, linewidth=0.25)
```

Sie uczy si znacznie mniej gwatownie ni bez normalizacji lub z momentem, jednak stabilizuje na bardzo wysokim poziomie.

```python
preds = net.predict(square_test_X)
plt.scatter(square_test_X, preds, linewidth=0.25)
```

Ksztat, ktrego uczy si sie nie przypomina tego oczekiwanego. Ze wzgldu na powolno oblicze nie udao mi si przetestowa wielu rnych architektur i parametrw uczenia.

# steps

## momentum

```python
np.random.seed(21)
net = mn.Net(1)
net.add(mn.DenseLayer(100, "sigmoid"))
net.add(mn.DenseLayer(1, "identity"))
net.kernel_init("xavier")

mse, iters, mses_history = net.backpropagate(steps_train_X, steps_train_Y, 
                               eta=0.000005, n_epochs=1000,
                               required_mse=3, batch_size=None, 
                               verbose=1,
                               validation_input=steps_test_X, 
                               validation_output=steps_test_Y, normalization_method="momentum"
                              )
```

```python
iters, mses = zip(*mses_history[10:])
plt.plot(iters, mses, linewidth=0.25)
```

MSE zaczyna bardzo wolno spada w okolicach MSE rwnego 50, jednak przez prawie 1000 epok od tego momentu wci powoli spada. Prawdopodobnie niewielka zmiana parametrw i uczenie sieci przez wicej epoko pozwolioby osign lepszy wynik. Jednak niestety ze wzgldu na wolne dziaanie algorytmu nie udao mi si przeprowadzi wicej eksperymentw.

```python
preds = net.predict(steps_test_X)
plt.scatter(steps_test_X, preds, linewidth=0.25)
```

Ksztat, ktrego uczy si sie przypomina ten oczekiwany.

## RMSProp

```python
np.random.seed(21)
net = mn.Net(1)
net.add(mn.DenseLayer(100, "sigmoid"))
net.add(mn.DenseLayer(1, "identity"))
net.kernel_init("xavier")

mse, iters, mses_history = net.backpropagate(steps_train_X, steps_train_Y, 
                               eta=0.1, n_epochs=700,
                               required_mse=3, batch_size=None, 
                               verbose=1,
                               validation_input=steps_test_X, 
                               validation_output=steps_test_Y, normalization_method="RMSProp"
                              )
```

```python
iters, mses = zip(*mses_history[10:])
plt.plot(iters, mses, linewidth=0.25)
```

MSE powoli spada od pewnego momentu. Znowu czas dziaania algorytmu nie pozwoli mi przeprowadzi uczenia na wikzej liczbie epok.

```python
preds = net.predict(steps_test_X)
plt.scatter(steps_test_X, preds, linewidth=0.25)
```

Ksztat przewidywany przez sie delikatnie przypomina ten oczekiwany. Sugeruje to, e metoda zaimplementowana jest poprawnie, jednak wymaga wicej epok i delikatnego tuningu hiperparamterw.

# Porwnanie szybkoci uczenia - bez normalizacji, z momentem, RMSProp

Uczenie bez normalizacji jest najbardziej gwatowne, najszybsze, ale jest te mniej stabilne (czciej wystpuj perturbacje wartoci MSE) ni uczenie z momentem. Uczenie z momentem pozwala czsto osign dokadniejsze dopasowania, jednak kosztem duszego dziaania algorytmu (wymagane wicej epok). Moe przedstawione powyej eksperymenty tego do koca nie udowadniaj, jednak przeprowadziem wczeniej te nie zaczone tu eksperymenty na square-simple, ktre pokazuj takie dziaanie.

Metoda RMSProp pozwala na najbardziej stabiln zbieno, jednak dziaa te najwolniej, przez co nie udao mi si w praktyce osign ni lepszych rezultatw wykorzystujc tlyko moc obliczeniow mojego komputera.
