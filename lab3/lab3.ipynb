```python
import pandas as pd
import numpy as np
import my_net as mn
import matplotlib.pyplot as plt
```

```python
square_train = pd.read_csv('../data/regression/square-large-training.csv', index_col=0)
steps_train = pd.read_csv('../data/regression/steps-large-training.csv', index_col=0)
square_test = pd.read_csv('../data/regression/square-large-test.csv', index_col=0)
steps_test = pd.read_csv('../data/regression/steps-large-test.csv', index_col=0)
square_test_X = square_test[['x']].values
square_test_Y = square_test[['y']].values
steps_test_X = steps_test[['x']].values
steps_test_Y = steps_test[['y']].values
square_train_X = square_train[['x']].values
square_train_Y = square_train[['y']].values
steps_train_X = steps_train[['x']].values
steps_train_Y = steps_train[['y']].values

multimodal_train = pd.read_csv('../data/regression/multimodal-large-training.csv')
multimodal_test = pd.read_csv('../data/regression/multimodal-large-test.csv')
multimodal_test_X = multimodal_test[['x']].values
multimodal_test_Y = multimodal_test[['y']].values
multimodal_train_X = multimodal_train[['x']].values
multimodal_train_Y = multimodal_train[['y']].values
```

# square

## momentum

```python
np.random.seed(21)
net = mn.Net(1)
net.add(mn.DenseLayer(25, "sigmoid"))
net.add(mn.DenseLayer(25, "sigmoid"))
net.add(mn.DenseLayer(25, "sigmoid"))
net.add(mn.DenseLayer(1, "identity"))
net.kernel_init("xavier")

mse, iters, mses_history = net.backpropagate(square_train_X, square_train_Y, 
                               eta=0.000005, n_epochs=3000,
                               required_mse=1, batch_size=None, 
                               verbose=1,
                               validation_input=square_test_X, 
                               validation_output=square_test_Y, normalization_method="momentum"
                              )
```

```python
iters, mses = zip(*mses_history[10:])
plt.plot(iters, mses, linewidth=0.25)
```

Sieć uczy się bardzo szybko, występują po drodze czasami pewne perturbacje, po których MSE znowu się stabilizuje.

```python
preds = net.predict(square_test_X)
plt.scatter(square_test_X, preds, linewidth=0.25)
```

Sieć jest w stanie nauczyć się w miarę podobnego kształtu do oczekiwanego, jednak MSE pozostaje na poziomie około 500. Inne nośniki zbioru testowego i treningowego utrudniają uzyskanie niskiego MSE na zbiorze testowym.

## RMSProp

```python
np.random.seed(21)
net = mn.Net(1)
net.add(mn.DenseLayer(25, "sigmoid"))
net.add(mn.DenseLayer(1, "identity"))
net.kernel_init("xavier")

mse, iters, mses_history = net.backpropagate(multimodal_train_X, multimodal_train_Y, 
                               eta=0.1, n_epochs=500,
                               required_mse=1, batch_size=None, 
                               verbose=1,
                               validation_input=multimodal_test_X, 
                               validation_output=multimodal_test_Y, normalization_method="RMSProp"
                              )
```

```python
iters, mses = zip(*mses_history[10:])
plt.plot(iters, mses, linewidth=0.25)
```

Sieć uczy się znacznie mniej gwałtownie niż bez normalizacji lub z momentem, jednak stabilizuje na bardzo wysokim poziomie.

```python
preds = net.predict(square_test_X)
plt.scatter(square_test_X, preds, linewidth=0.25)
```

Kształt, którego uczy się sieć nie przypomina tego oczekiwanego. Ze względu na powolność obliczeń nie udało mi się przetestować wielu różnych architektur i parametrów uczenia.

# steps

## momentum

```python
np.random.seed(21)
net = mn.Net(1)
net.add(mn.DenseLayer(100, "sigmoid"))
net.add(mn.DenseLayer(1, "identity"))
net.kernel_init("xavier")

mse, iters, mses_history = net.backpropagate(steps_train_X, steps_train_Y, 
                               eta=0.000005, n_epochs=1000,
                               required_mse=3, batch_size=None, 
                               verbose=1,
                               validation_input=steps_test_X, 
                               validation_output=steps_test_Y, normalization_method="momentum"
                              )
```

```python
iters, mses = zip(*mses_history[10:])
plt.plot(iters, mses, linewidth=0.25)
```

MSE zaczyna bardzo wolno spadać w okolicach MSE równego 50, jednak przez prawie 1000 epok od tego momentu wciąż powoli spada. Prawdopodobnie niewielka zmiana parametrów i uczenie sieci przez więcej epoko pozwoliłoby osiągnąć lepszy wynik. Jednak niestety ze względu na wolne działanie algorytmu nie udało mi się przeprowadzić więcej eksperymentów.

```python
preds = net.predict(steps_test_X)
plt.scatter(steps_test_X, preds, linewidth=0.25)
```

Kształt, którego uczy się sieć przypomina ten oczekiwany.

## RMSProp

```python
np.random.seed(21)
net = mn.Net(1)
net.add(mn.DenseLayer(100, "sigmoid"))
net.add(mn.DenseLayer(1, "identity"))
net.kernel_init("xavier")

mse, iters, mses_history = net.backpropagate(steps_train_X, steps_train_Y, 
                               eta=0.1, n_epochs=700,
                               required_mse=3, batch_size=None, 
                               verbose=1,
                               validation_input=steps_test_X, 
                               validation_output=steps_test_Y, normalization_method="RMSProp"
                              )
```

```python
iters, mses = zip(*mses_history[10:])
plt.plot(iters, mses, linewidth=0.25)
```

MSE powoli spada od pewnego momentu. Znowu czas działania algorytmu nie pozwolił mi przeprowadzić uczenia na więkzej liczbie epok.

```python
preds = net.predict(steps_test_X)
plt.scatter(steps_test_X, preds, linewidth=0.25)
```

Kształt przewidywany przez sieć delikatnie przypomina ten oczekiwany. Sugeruje to, że metoda zaimplementowana jest poprawnie, jednak wymaga więcej epok i delikatnego tuningu hiperparamterów.

# Porównanie szybkości uczenia - bez normalizacji, z momentem, RMSProp

Uczenie bez normalizacji jest najbardziej gwałtowne, najszybsze, ale jest też mniej stabilne (częściej występują perturbacje wartości MSE) niż uczenie z momentem. Uczenie z momentem pozwala często osiągnąć dokładniejsze dopasowania, jednak kosztem dłuższego działania algorytmu (wymagane więcej epok). Może przedstawione powyżej eksperymenty tego do końca nie udowadniają, jednak przeprowadziłem wcześniej też nie załączone tu eksperymenty na square-simple, które pokazują takie działanie.

Metoda RMSProp pozwala na najbardziej stabilną zbieżność, jednak działa też najwolniej, przez co nie udało mi się w praktyce osiągnąć nią lepszych rezultatów wykorzystując tlyko moc obliczeniową mojego komputera.
